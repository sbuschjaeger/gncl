```python
import plotly.express as px
from skimage import io
from PIL import Image 

def display_thumbnail(img, size = 512):
    for_display = img.copy()
    maxsize = (size, size)
    for_display.thumbnail(maxsize)
    display(for_display)

image_path = '/data/s1/buschjae/PNEUMONIA/chest_xray/train/NORMAL/NORMAL2-IM-1423-0001.jpeg'
img = Image.open(image_path) 
width, height = img.size 

print("W / H: {} / {}".format(width, height))  
display_thumbnail(img, 256)
```
