```python
import plotly.express as px
from skimage import io
from PIL import Image 

def display_thumbnail(img, size = 512):
    for_display = img.copy()
    maxsize = (size, size)
    for_display.thumbnail(maxsize)
    display(for_display)

image_path = '/data/d3/buschjae/gncl/cowc/Columbus_CSUAV_AFRL/train/23.01712.02080.060.png'
img = Image.open(image_path) 
width, height = img.size 
print(img.size)
print(img.mode)
#img.shape
#print("W / H: {} / {}".format(width, height))  
display_thumbnail(img, 256)

wc = 196
hc = 196
cropped = img.crop((width//2 - wc//2, height//2 - hc//2, width//2 + wc//2, height//2 + hc//2))
cropped
display_thumbnail(cropped, 128)
```
